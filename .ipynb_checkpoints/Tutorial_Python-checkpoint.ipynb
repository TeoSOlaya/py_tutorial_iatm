```python
# Para las funciones 
def potencia(a,b):
    output = a ** b
    return [output,a] # si quiero devolver varias variables, las regreso en forma de vector
```

```python
potencia(2,8)
```

```python
[potencia(x,2) for x in range(0,10,2)] #€quivalente a matlab 0:9
```

```python
#Test estadístico

from scipy.stats import ttest_ind

setosa = iris[iris['species'] == 'setosa']
virginica = iris[iris['species'] == 'virginica']

print(ttest_ind(setosa['sepal_length'], virginica['sepal_length']))
print(ttest_ind(setosa['sepal_width'], virginica['sepal_width']))
```

```python
# Anatomía
# First, we fetch and load the default anatomical image available in Nilearn datasets module
from nilearn import datasets

#default MNI152 template
anatomica1_image = datasets.load_mni152_template()
# anatomica1_image = "/Users/gloriaalicia.... según la librería que tenga"

# Then, import plotting module from nilearn to easily visualize the loaded image
from nilearn import plotting

%matplotlib inline
plotting.plot_anat(anatomica1_image, title = 'MNI template')
```

```python
# Funcional
# Fetch t-contrasts maps from Nilearn datasets module
localizer_dataset = datasets.fetch_localizer_button_task(
    n_subjects = 1,
    get_anats = True)

#Anatomical image filename
localizer_anat_filename = localizer_dataset.anats[0]
print (localizer_anat_filename)

# T-contrast map filename
localizer_tmap_filename = localizer_dataset.tmaps[0]
print (localizer_tmap_filename)
```

```python
#visualizar t-map with 'thresold = 3' based on its statical intensities 
plotting.plot_stat_map (localizer_tmap_filename, 
                        bg_img = localizer_anat_filename, 
                        threshold = 3,
                        title = "Statistical map with threshold = 3")
```

```python
# Visualización with plot_abs = Fase & threshold = 3
plotting.plot_glass_brain (localizer_tmap_filename, 
                        threshold = 3, plot_abs = False, colorbar = True,
                        title = "Statistical map with plot_abs=False & threshold = 3")
```

```python
# Showing percentile threshold image
plotting.plot_stat_map (localizer_tmap_filename, display_mode='z', threshold=3, 
                        cut_coords = 5, colorbar=True)
```
